```python


str07 = input("Enter a sentence : ")
count = 0
alpha = ["Q","W","E","R","T","Y","U","I","O","P","A","S","D","F","G","H","J","K","L","Z","X","C","V","B","N","M"]
for i in range(len(str07)):
    if str07[i] in alpha:
      count = count + 1
print("The given sentence has",count,"upper-case letters.")
```

9. Create a program that checks if a given string is a palindrome.



```python
str05 = input("Enter a sentence : ")
a = []
for i in str05.lower():
  a.append(i)

for j in range(1000):
  for i in a:
    if i == " ":
      a.remove(" ")
    if i == ",":
      a.remove(",")
    if i == ".":
      a.remove(".")
    if i == "'":
      a.remove("'")
    if i == "?":
      a.remove("?")

if a == a[::-1]:
  print("The given sentence is a palindrome.")
else:
  print("The given sentence is not a palindrome.")


```

12. Write a program that finds the longest word in a given sentence.



```python
str04 = input("Enter a sentence : ")
word = str04.split(" ")
longest = [word[0]]

for i in range(1,len(word)):
  if len(word[i]) > len(longest[0]):
    del longest[0]
    longest = [word[i]]
  elif len(word[i]) == len(longest[0]):
    longest.append(word[i])

if len(longest) == 1:
  print("The longest word is : ",longest[0])
else:
  print('The longest words are : ', ", ".join(longest))

```
